```python
import numpy as np
import matplotlib.pyplot as plt


# Construct sigma n= 200
rho_1 = 0.8
rho_2 = 0.7
rho_3 = 0.3
Sigma = np.full((200, 200), rho_3, dtype=float)

# set block1 and block2
Sigma[:100, :100] = rho_1       # first block
Sigma[100:, 100:] = rho_2       # second block

# set diagonal to 1
np.fill_diagonal(Sigma, 1.0)

# exp value
mu = np.zeros(200)

# get C with Cholesky decomposition
C = np.linalg.cholesky(Sigma)

# simulate N=10 000 realizations of X ~ N(mu, Sigma)
X_list = []
for _ in range(10000):
    # vector with n independend standard normal variables
    Y = np.random.normal(0, 1, 200)

    # simulate X
    X = mu + C @ Y
    X_list.append(X)

X_array = np.array(X_list)
print("Shape of X_array:", X_array.shape)
```
